Given a list of numbers, find the sum and average.


```python
# 1. Given a list of numbers, find the sum and average.
def sum_and_average(numbers):
    """Calculates the sum and average of a list of numbers."""
    if not numbers:  # Handle empty list case
        return 0, 0
    total = sum(numbers)
    average = total / len(numbers)
    return total, average

numbers = [10, 20, 30, 40, 50]
sum_val, avg_val = sum_and_average(numbers)
print("Sum:", sum_val)
print("Average:", avg_val)
```

Create a program that takes a temperature in Celsius and converts it to Kelvin.


```python
# 2. Create a program that takes a temperature in Celsius and converts it to Kelvin.
def celsius_to_kelvin(celsius):
    """Converts Celsius to Kelvin."""
    kelvin = celsius + 273.15
    return kelvin

celsius = float(input("Enter temperature in Celsius: "))
kelvin = celsius_to_kelvin(celsius)
print("Temperature in Kelvin:", kelvin)

```

Implement a program that checks if a given string is a palindrome.


```python

# 3. Implement a program that checks if a given string is a palindrome.
def is_palindrome(text):
    """Checks if a string is a palindrome (case-insensitive, ignores spaces)."""
    processed_text = text.lower().replace(" ", "")
    reversed_text = processed_text[::-1]
    return processed_text == reversed_text

text = input("Enter a string: ")
if is_palindrome(text):
    print(text, "is a palindrome.")
else:
    print(text, "is not a palindrome.")
```

Create a function to reverse a given string.


```python
# 4. Create a function to reverse a given string.
def reverse_string(text):
    """Reverses a string."""
    return text[::-1]

text = input("Enter a string: ")
reversed_text = reverse_string(text)
print("Reversed string:", reversed_text)

```

Given a list of names, concatenate them into a single string separated by spaces.


```python
# 5. Given a list of names, concatenate them into a single string separated by spaces.
def concatenate_names(names):
    """Concatenates a list of names into a single string."""
    return " ".join(names)

names = ["Alice", "Bob", "Charlie"]
concatenated_string = concatenate_names(names)
print("Concatenated string:", concatenated_string)
```

Write a Python program to check if a given string is a pangram (contains all letters of the alphabet).


```python

# 6. Write a Python program to check if a given string is a pangram (contains all letters of the alphabet).
import string

def is_pangram(text):
    """Checks if a string is a pangram (case-insensitive)."""
    alphabet = set(string.ascii_lowercase)
    processed_text = text.lower()
    return alphabet.issubset(set(processed_text))

text = input("Enter a string: ")
if is_pangram(text):
    print(text, "is a pangram.")
else:
    print(text, "is not a pangram.")

```

Calculate the area and circumference of a circle given its radius.


```python
# 7. Calculate the area and circumference of a circle given its radius.
import math

def circle_calculations(radius):
    """Calculates the area and circumference of a circle."""
    area = math.pi * radius**2
    circumference = 2 * math.pi * radius
    return area, circumference

radius = float(input("Enter the radius of the circle: "))
area, circumference = circle_calculations(radius)
print("Area:", area)
print("Circumference:", circumference)
```

Implement a program that converts a given number of minutes into hours and minutes.


```python
# 8. Implement a program that converts a given number of minutes into hours and minutes.
def convert_minutes(minutes):
    """Converts minutes into hours and remaining minutes."""
    hours = minutes // 60
    remaining_minutes = minutes % 60
    return hours, remaining_minutes

minutes = int(input("Enter the number of minutes: "))
hours, remaining_minutes = convert_minutes(minutes)
print(hours, "hours and", remaining_minutes, "minutes")
```

Create a function to count the number of vowels in a given string.


```python

# 9. Create a function to count the number of vowels in a given string.
def count_vowels(text):
    """Counts the number of vowels in a string (case-insensitive)."""
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

text = input("Enter a string: ")
vowel_count = count_vowels(text)
print("Number of vowels:", vowel_count)
```

Write a program to check if a number is prime.


```python

# 10. Write a program to check if a number is prime.
def is_prime(number):
    """Checks if a number is prime."""
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

number = int(input("Enter a number: "))
if is_prime(number):
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```
