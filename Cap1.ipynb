Write a Python program to calculate the area of a rectangle given its length and width.


```python
# 1. Calculate the area of a rectangle
def calculate_rectangle_area(length, width):
    """Calculates the area of a rectangle."""
    area = length * width
    return area

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
rectangle_area = calculate_rectangle_area(length, width)
print("The area of the rectangle is:", rectangle_area)

```

Write a program to check if a number is even or odd.


```python
# 3. Check if a number is even or odd
def check_even_odd(number):
    """Checks if a number is even or odd."""
    if number % 2 == 0:
        print(number, "is even.")
    else:
        print(number, "is odd.")

number = int(input("Enter a number: "))
check_even_odd(number)
```

Given a list of numbers, find the maximum and minimum values.


```python
# 4. Find the maximum and minimum values in a list
def find_max_min(numbers):
    """Finds the maximum and minimum values in a list."""
    maximum = max(numbers)
    minimum = min(numbers)
    return maximum, minimum

numbers = [5, 2, 9, 1, 5, 6]
max_value, min_value = find_max_min(numbers)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

Create a Python function to check if a given string is a palindrome.


```python
# 5. Check if a string is a palindrome
def is_palindrome(text):
    """Checks if a string is a palindrome."""
    processed_text = text.lower().replace(" ", "")  # Remove spaces and lowercase
    reversed_text = processed_text[::-1]  # Reverse the string
    if processed_text == reversed_text:
        return True
    else:
        return False

text = input("Enter a string: ")
if is_palindrome(text):
    print(text, "is a palindrome.")
else:
    print(text, "is not a palindrome.")

```

Calculate the compound interest for a given principal amount, interest rate, and time period.


```python
# 6. Calculate compound interest
def calculate_compound_interest(principal, rate, time):
    """Calculates compound interest."""
    amount = principal * (1 + rate / 100) ** time
    interest = amount - principal
    return interest

principal = float(input("Enter principal amount: "))
rate = float(input("Enter annual interest rate: "))
time = float(input("Enter time period (years): "))
compound_interest = calculate_compound_interest(principal, rate, time)
print("Compound interest:", compound_interest)
```

Write a program that converts a given number of days into years, weeks, and days.


```python
# 7. Convert days into years, weeks, and days
def convert_days(total_days):
    """Converts days into years, weeks, and remaining days."""
    years = total_days // 365
    remaining_days = total_days % 365
    weeks = remaining_days // 7
    days = remaining_days % 7
    return years, weeks, days

total_days = int(input("Enter the number of days: "))
years, weeks, days = convert_days(total_days)
print(years, "years,", weeks, "weeks, and", days, "days")
```

Given a list of integers, find the sum of all positive numbers.


```python
# 8. Sum of positive numbers in a list
def sum_positive_numbers(numbers):
    """Calculates the sum of positive numbers in a list."""
    positive_sum = 0
    for num in numbers:
        if num > 0:
            positive_sum += num
    return positive_sum

numbers = [1, -2, 3, -4, 5]
positive_sum = sum_positive_numbers(numbers)
print("Sum of positive numbers:", positive_sum)
```

Create a program that takes a sentence as input and counts the number of words in it.


```python
# 9. Count words in a sentence
def count_words(sentence):
    """Counts the number of words in a sentence."""
    words = sentence.split()
    return len(words)

sentence = input("Enter a sentence: ")
word_count = count_words(sentence)
print("Number of words:", word_count)
```

Implement a program that swaps the values of two variables.

```python
# 10. Swap values of two variables
def swap_variables(a, b):
    """Swaps the values of two variables."""
    a, b = b, a
    return a, b

x = 10
y = 5
print("Before swapping: x =", x, "y =", y)
x, y = swap_variables(x, y)
print("After swapping: x =", x, "y =", y)
```
