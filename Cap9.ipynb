Create a class to represent a Student with attributes like name, age, and grades.


```python
import csv
import json

# 1. Create a class to represent a Student with attributes like name, age, and grades.
class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def get_average_grade(self):
        if self.grades:
            return sum(self.grades) / len(self.grades)
        return 0

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Average Grade: {self.get_average_grade()}"

# Example Usage:
student1 = Student("Alice", 20, [85, 90, 78, 92])
print(student1)
```

Given a CSV file with employee details (name, position, salary), create a class to represent an Employee.


```python


# 2. Given a CSV file with employee details (name, position, salary), create a class to represent an Employee.
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def __str__(self):
        return f"Name: {self.name}, Position: {self.position}, Salary: {self.salary}"

def create_employees_from_csv(csv_file):
    employees = []
    try:
        with open(csv_file, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                employees.append(Employee(row['name'], row['position'], float(row['salary'])))
    except FileNotFoundError:
        print(f"Error: File '{csv_file}' not found.")
    except KeyError:
        print("Error: CSV file missing required columns (name, position, salary).")
    return employees

# Example Usage:
employees = create_employees_from_csv('employees.csv')
for emp in employees:
    print(emp)
```

Implement a program that simulates a basic bank account using a BankAccount class.


```python


# 3. Implement a program that simulates a basic bank account using a BankAccount class.
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds or invalid amount.")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Balance: ${self.balance}"

# Example Usage:
account1 = BankAccount("1234567890", 1000)
print(account1)
account1.deposit(500)
account1.withdraw(200)
account1.withdraw(1500)
```

Write a Python program that uses a Rectangle class to calculate the area and perimeter of a rectangle.


```python


# 4. Write a Python program that uses a Rectangle class to calculate the area and perimeter of a rectangle.
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

    def __str__(self):
        return f"Length: {self.length}, Width: {self.width}, Area: {self.calculate_area()}, Perimeter: {self.calculate_perimeter()}"

# Example Usage:
rect1 = Rectangle(10, 5)
print(rect1)
```

Given a JSON file with customer data, create a Customer class to store and manipulate the data.


```python


# 6. Given a JSON file with customer data, create a Customer class to store and manipulate the data.
class Customer:
    def __init__(self, customer_id, name, email, purchases=[]):
        self.customer_id = customer_id
        self.name = name
        self.email = email
        self.purchases = purchases  # List of purchase amounts

    def add_purchase(self, amount):
        if amount > 0:
            self.purchases.append(amount)

    def get_total_purchases(self):
        return sum(self.purchases)

    def __str__(self):
        return f"ID: {self.customer_id}, Name: {self.name}, Email: {self.email}, Total Purchases: ${self.get_total_purchases()}"

def create_customer_from_json(json_file):
    try:
        with open(json_file, 'r') as file:
            data = json.load(file)
            return Customer(data['customer_id'], data['name'], data['email'], data.get('purchases', []))
    except FileNotFoundError:
        print(f"Error: File '{json_file}' not found.")
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON format in '{json_file}'.")
    return None

# Example Usage:
customer = create_customer_from_json('customer_data.json')
if customer:
    print(customer)
    customer.add_purchase(100)
    customer.add_purchase(25.50)
    print(customer)
```

Implement a program that uses a Circle class to calculate the area and circumference of multiple circles.


```python

# 8. Implement a program that uses a Circle class to calculate the area and circumference of multiple circles.
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def calculate_circumference(self):
        return 2 * math.pi * self.radius

    def __str__(self):
        return f"Radius: {self.radius}, Area: {self.calculate_area():.2f}, Circumference: {self.calculate_circumference():.2f}"

# Example Usage:
circles = [Circle(5), Circle(10), Circle(7.5)]
for circle in circles:
    print(circle)
```

Given a CSV file with product details (name, price, quantity), create a Product class to manage the data.


```python


# 9. Given a CSV file with product details (name, price, quantity), create a Product class to manage the data.
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_total_value(self):
        return self.price * self.quantity

    def __str__(self):
        return f"Name: {self.name}, Price: ${self.price:.2f}, Quantity: {self.quantity}, Total Value: ${self.calculate_total_value():.2f}"

def create_products_from_csv(csv_file):
    products = []
    try:
        with open(csv_file, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                products.append(Product(row['name'], float(row['price']), int(row['quantity'])))
    except FileNotFoundError:
        print(f"Error: File '{csv_file}' not found.")
    except KeyError:
        print("Error: CSV file missing required columns (name, price, quantity).")
    return products

# Example Usage:
products = create_products_from_csv('products.csv')
for prod in products:
    print(prod)
```
