Given a list of numbers, create a NumPy array and find the sum and product of its elements.


```python
import numpy as np

# 1. Given a list of numbers, create a NumPy array and find the sum and product of its elements.
def sum_and_product(numbers):
    """
    Creates a NumPy array from a list and calculates the sum and product of its elements.

    Args:
        numbers (list): A list of numbers.

    Returns:
        tuple: A tuple containing the sum and product of the array elements.
    """
    numpy_array = np.array(numbers)
    array_sum = np.sum(numpy_array)
    array_product = np.prod(numpy_array)
    return array_sum, array_product

numbers_list = [1, 2, 3, 4, 5]
sum_result, product_result = sum_and_product(numbers_list)
print(f"1. List: {numbers_list}")
print(f"   Sum: {sum_result}")
print(f"   Product: {product_result}\n")
```

Write a Python program that uses NumPy to find the mean, median, and standard deviation of a dataset.


```python



# 3. Write a Python program that uses NumPy to find the mean, median, and standard deviation of a dataset.
def calculate_stats(data):
    """
    Calculates the mean, median, and standard deviation of a NumPy array.

    Args:
        data (np.ndarray): A NumPy array of numerical data.

    Returns:
        tuple: A tuple containing the mean, median, and standard deviation.
    """
    mean_value = np.mean(data)
    median_value = np.median(data)
    std_dev_value = np.std(data)
    return mean_value, median_value, std_dev_value

dataset = np.array([10, 15, 20, 12, 18, 25])
mean_val, median_val, std_dev_val = calculate_stats(dataset)
print(f"3. Dataset: {dataset}")
print(f"   Mean: {mean_val}")
print(f"   Median: {median_val}")
print(f"   Standard Deviation: {std_dev_val}\n")
```

Create a function that takes a list of numbers and returns the NumPy array sorted in ascending order.


```python

# 4. Create a function that takes a list of numbers and returns the NumPy array sorted in ascending order.
def sort_array_ascending(numbers):
    """
    Takes a list of numbers and returns a NumPy array sorted in ascending order.

    Args:
        numbers (list): A list of numbers.

    Returns:
        np.ndarray: A NumPy array sorted in ascending order.
    """
    numpy_array = np.array(numbers)
    sorted_array = np.sort(numpy_array)
    return sorted_array

unsorted_list = [5, 1, 9, 3, 7]
sorted_array_result = sort_array_ascending(unsorted_list)
print(f"4. Unsorted list: {unsorted_list}")
print(f"   Sorted NumPy array: {sorted_array_result}\n")
```

Given a list of lists, create a 2D NumPy array and find the sum of elements in each row and column.


```python


# 5. Given a list of lists, create a 2D NumPy array and find the sum of elements in each row and column.
def row_and_column_sums(list_of_lists):
    """
    Creates a 2D NumPy array from a list of lists and finds the sum of elements in each row and column.

    Args:
        list_of_lists (list): A list of lists (representing rows).

    Returns:
        tuple: A tuple containing the array, row sums, and column sums.
    """
    numpy_2d_array = np.array(list_of_lists)
    row_sums = np.sum(numpy_2d_array, axis=1)
    column_sums = np.sum(numpy_2d_array, axis=0)
    return numpy_2d_array, row_sums, column_sums

data_2d = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
array_2d, row_sums_result, column_sums_result = row_and_column_sums(data_2d)
print(f"5. List of lists:\n{data_2d}")
print(f"   2D NumPy array:\n{array_2d}")
print(f"   Row sums: {row_sums_result}")
print(f"   Column sums: {column_sums_result}\n")
```

Write a function that takes a NumPy array and returns a new array with all elements squared.


```python

# 7. Write a function that takes a NumPy array and returns a new array with all elements squared.
def square_array_elements(numpy_array):
    """
    Takes a NumPy array and returns a new array with all elements squared.

    Args:
        numpy_array (np.ndarray): The input NumPy array.

    Returns:
        np.ndarray: A new NumPy array with elements squared.
    """
    squared_array = numpy_array ** 2
    return squared_array

sample_array = np.array([1, 2, 3, 4])
squared_array_result = square_array_elements(sample_array)
print(f"7. Original NumPy array: {sample_array}")
print(f"   Squared NumPy array: {squared_array_result}\n")
```

Create a program that uses NumPy to calculate the inverse of a 2x2 matrix.


```python
# 9. Create a program that uses NumPy to calculate the inverse of a 2x2 matrix.
matrix_2x2 = np.array([[1, 2], [3, 4]])
try:
    inverse_matrix = np.linalg.inv(matrix_2x2)
    print("9. 2x2 Matrix:\n", matrix_2x2)
    print("\n   Inverse Matrix:\n", inverse_matrix, "\n")
except np.linalg.LinAlgError:
    print("9. The matrix is singular (non-invertible).\n")
```

Implement a function that takes a NumPy array and returns the transpose of the array.

```python
# 10. Implement a function that takes a NumPy array and returns the transpose of the array.
def transpose_array(numpy_array):
    """
    Takes a NumPy array and returns its transpose.

    Args:
        numpy_array (np.ndarray): The input NumPy array.

    Returns:
        np.ndarray: The transpose of the input array.
    """
    transposed_array = numpy_array.T
    return transposed_array

sample_matrix = np.array([[1, 2], [3, 4], [5, 6]])
transposed_matrix = transpose_array(sample_matrix)
print("10. Original NumPy array:\n", sample_matrix)
print("\n    Transpose of the array:\n", transposed_matrix)
```
