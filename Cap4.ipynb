Given a list of numbers, create a function to find the sum of all positive numbers.  


```python
# 1. Given a list of numbers, create a function to find the sum of all positive numbers.
def sum_positive_numbers(numbers):
    """Calculates the sum of positive numbers in a list."""
    positive_sum = 0
    for num in numbers:
        if num > 0:
            positive_sum += num
    return positive_sum

numbers = [1, -2, 3, -4, 5]
positive_sum = sum_positive_numbers(numbers)
print("Sum of positive numbers:", positive_sum)

```

Write a Python function to check if a given string is a palindrome.  


```python
# 2. Write a Python function to check if a given string is a palindrome.
def is_palindrome(text):
    """Checks if a string is a palindrome (case-insensitive, ignores spaces)."""
    processed_text = text.lower().replace(" ", "")
    reversed_text = processed_text[::-1]
    return processed_text == reversed_text

text = input("Enter a string: ")
if is_palindrome(text):
    print(text, "is a palindrome.")
else:
    print(text, "is not a palindrome.")
```

Implement a function that returns the factorial of a given number using recursion.  


```python
# 3. Implement a function that returns the factorial of a given number using recursion.
def factorial_recursive(n):
    """Calculates the factorial of a number using recursion."""
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

number = int(input("Enter a number: "))
fact = factorial_recursive(number)
print("Factorial:", fact)
```

Create a function to find the square of each element in a given list.  


```python

# 4. Create a function to find the square of each element in a given list.
def square_elements(numbers):
    """Returns a new list with the square of each element."""
    squared_numbers = [num ** 2 for num in numbers]
    return squared_numbers

numbers = [1, 2, 3, 4, 5]
squared_nums = square_elements(numbers)
print("Squared numbers:", squared_nums)
```

Write a function to check if a number is even or odd and return "Even" or "Odd" accordingly.  


```python
# 5. Write a function to check if a number is even or odd and return "Even" or "Odd" accordingly.
def check_even_odd(number):
    """Checks if a number is even or odd and returns "Even" or "Odd"."""
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

number = int(input("Enter a number: "))
result = check_even_odd(number)
print(number, "is", result)
```

Calculate the area of a triangle given its base and height using a function.  


```python
# 6. Calculate the area of a triangle given its base and height using a function.
def triangle_area(base, height):
    """Calculates the area of a triangle."""
    area = 0.5 * base * height
    return area

base = float(input("Enter the base of the triangle: "))
height = float(input("Enter the height of the triangle: "))
area = triangle_area(base, height)
print("Area of the triangle:", area)

```

Create a function that takes a list of strings and returns the list sorted alphabetically.  


```python

# 7. Create a function that takes a list of strings and returns the list sorted alphabetically.
def sort_strings_alphabetically(strings):
    """Sorts a list of strings alphabetically."""
    strings.sort()  # Sorts the list in place
    return strings

strings = ["banana", "apple", "cherry", "date"]
sorted_strings = sort_strings_alphabetically(strings)
print("Sorted strings:", sorted_strings)
```

Write a function that takes two lists and returns their intersection (common elements).  


```python
# 8. Write a function that takes two lists and returns their intersection (common elements).
def find_intersection(list1, list2):
    """Finds the intersection of two lists."""
    intersection = [item for item in list1 if item in list2]
    return intersection

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common_elements = find_intersection(list1, list2)
print("Common elements:", common_elements)
```

Implement a function to check if a given year is a leap year or not.  


```python
# 9. Implement a function to check if a given year is a leap year or not.
def is_leap_year(year):
    """Checks if a given year is a leap year."""
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Enter a year: "))
if is_leap_year(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

Create a function that takes a number as input and prints its multiplication table.  


```python
# 10. Create a function that takes a number as input and prints its multiplication table.
def multiplication_table(number):
    """Prints the multiplication table of a given number."""
    for i in range(1, 11):
        print(number, "x", i, "=", number * i)

number = int(input("Enter a number: "))
multiplication_table(number)
```
