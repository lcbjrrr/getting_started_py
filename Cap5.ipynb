Given two lists of numbers, concatenate them into a single list.


```python

# 1. Given two lists of numbers, concatenate them into a single list.
def concatenate_lists(list1, list2):
    """Concatenates two lists into a single list."""
    return list1 + list2

list1 = [1, 2, 3]
list2 = [4, 5, 6]
concatenated_list = concatenate_lists(list1, list2)
print("Concatenated list:", concatenated_list)
```

Write a program that finds the largest and smallest elements in a list.


```python

# 2. Write a program that finds the largest and smallest elements in a list.
def find_largest_smallest(numbers):
    """Finds the largest and smallest elements in a list."""
    if not numbers:  # Handle empty list case
        return None, None
    largest = max(numbers)
    smallest = min(numbers)
    return largest, smallest

numbers = [5, 2, 9, 1, 5, 6]
largest_num, smallest_num = find_largest_smallest(numbers)
print("Largest:", largest_num)
print("Smallest:", smallest_num)
```

Implement a function that takes a list of numbers and returns a new list with the squared values.


```python
# 3. Implement a function that takes a list of numbers and returns a new list with the squared values.
def square_elements(numbers):
    """Returns a new list with the square of each element."""
    squared_numbers = [num ** 2 for num in numbers]
    return squared_numbers

numbers = [1, 2, 3, 4, 5]
squared_nums = square_elements(numbers)
print("Squared numbers:", squared_nums)
```

Create a program that finds the common elements between two lists and stores them in a new list.


```python
# 4. Create a program that finds the common elements between two lists and stores them in a new list.
def find_common_elements(list1, list2):
    """Finds the common elements between two lists."""
    common_elements = [item for item in list1 if item in list2]
    return common_elements

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common_nums = find_common_elements(list1, list2)
print("Common elements:", common_nums)
```

Given a list of words, find the word with the maximum length and its length.


```python
# 5. Given a list of words, find the word with the maximum length and its length.
def find_longest_word(words):
    """Finds the longest word in a list and its length."""
    if not words:  # Handle empty list case
        return None, 0
    max_word = ""
    max_len = 0
    for word in words:
        if len(word) > max_len:
            max_len = len(word)
            max_word = word
    return max_word, max_len

words = ["apple", "banana", "kiwi", "orange"]
longest_word, length = find_longest_word(words)
print("Longest word:", longest_word)
print("Length:", length)
```

Write a Python program to count the occurrences of each element in a given list.


```python
# 6. Write a Python program to count the occurrences of each element in a given list.
def count_element_occurrences(elements):
    """Counts the occurrences of each element in a list."""
    counts = {}
    for element in elements:
        if element in counts:
            counts[element] += 1
        else:
            counts[element] = 1
    return counts

elements = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
occurrences = count_element_occurrences(elements)
print("Occurrences:", occurrences)
```

Given a list of names, remove all duplicate names and print the unique names.


```python
# 7. Given a list of names, remove all duplicate names and print the unique names.
def get_unique_names(names):
    """Removes duplicate names from a list, preserving order."""
    unique_names = []
    seen = set()  # Use a set to keep track of seen names
    for name in names:
        if name not in seen:
            unique_names.append(name)
            seen.add(name)
    return unique_names

names = ["Alice", "Bob", "Alice", "Charlie", "Bob", "David"]
unique_names = get_unique_names(names)
print("Unique names:", unique_names)
```

Create a function that takes a list of strings and returns the list sorted by the length of the strings.


```python

# 8. Create a function that takes a list of strings and returns the list sorted by the length of the strings.
def sort_by_length(strings):
    """Sorts a list of strings by their length."""
    return sorted(strings, key=len)

strings = ["apple", "banana", "kiwi", "orange"]
sorted_strings = sort_by_length(strings)
print("Sorted by length:", sorted_strings)
```

Write a program that checks if a given list is sorted in ascending order.


```python
# 9. Write a program that checks if a given list is sorted in ascending order.
def is_sorted_ascending(numbers):
    """Checks if a list of numbers is sorted in ascending order."""
    for i in range(len(numbers) - 1):
        if numbers[i] > numbers[i + 1]:
            return False
    return True

numbers1 = [1, 2, 3, 4, 5]
numbers2 = [5, 2, 8, 1, 9]
print("Is sorted (ascending):", is_sorted_ascending(numbers1))
print("Is sorted (ascending):", is_sorted_ascending(numbers2))
```

Implement a function that takes two lists and returns their union (all unique elements from both lists).

```python
# 10. Implement a function that takes two lists and returns their union (all unique elements from both lists).
def find_union(list1, list2):
    """Finds the union of two lists (all unique elements)."""
    union_set = set(list1) | set(list2)
    return list(union_set)

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
union_list = find_union(list1, list2)
print("Union:", union_list)
```
